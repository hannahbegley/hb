```python
"""code to remove duplicates from remaining lists in data"""
"""set a list of columns to loop through and de-dup and initialize a counter to loop through the list of columns"""
de_dup_list = ["affiliations", "contributors", "publishers", "tags"]
s = 0

while s < len(de_dup_list):
    """select a column from the de_dup_list of columns"""
    selected_column = de_dup_list[s]
    print selected_column 
    """to confirm selected column"""
    """set counter for row within that column"""
    i = 0
    """set new_col to a blank column- this is where de-duped row values will be appended"""
    new_col = []
    
    while i < len(testing[selected_column]):
#     while i < 10:
        #fill in blank cells so won't throw error
        testing[selected_column] = testing[selected_column].fillna("[]")
        """select a single row"""
        row = testing[selected_column][i]    
        row = [str(x) for x in row.split(',')]  
    #     print "original values:", row, len(row)
        new_list = []
        for item in row:
            """remove lingering json list formatting and convert unique values into list-
            this is instead of translating the json list into a dataframe as done above"""
            item = item.strip("[").strip("]").strip() 
            if item not in new_list:
                new_list.append(item)
            new_list = list(set(new_list))
        new_col.append(new_list)
    #     print "de-duplicated values: ", new_list, len(new_list)
        i += 1
        """assign cleaned up values to original dataframe """
    print len(new_col)
    testing[selected_column] = new_col
#         print new_col
    s +=1
```

```python
#write the above dataframe, with cleaned up lists for contributors, affiliations, publishers, and tags, to a new excel doc
# testing.to_excel('SHARE_cleaned_lists.xlsx', index=False)
testing.to_csv('SHARE_cleaned_lists.csv', index=False)
```

```python
#and then read in this document to analyze
# df = pd.read_excel('SHARE_cleaned_lists.xlsx', index=False)
df = pd.read_csv('SHARE_cleaned_lists.csv')
```

```python
df.head()
```

```python
"""write all affiliations into a single list to review-
leave the quotations on to make it easy to distinguish what individual values in the raw data were"""
affiliations_all = []
for row in df['affiliations']:
    row = row.strip("[").strip("]").strip() 
    row = [str(x) for x in row.split(',')]
    for item in row:
        affiliations_all.append(item)
```

```python
"""write all contributors into a single list to review"""
contributors_all = []
for row in df['contributors']:
    row = row.strip("[").strip("]").strip() 
    row = [str(x) for x in row.split(',')]
    for item in row:
        contributors_all.append(item)
```

```python
"""write both to their own dataframe, count the number of occurrences, and save to csv file containing the term
and number of occurrences"""
affiliations_eval = pd.DataFrame(np.array(affiliations_all))
affiliations_eval.columns = ['affiliations']
affiliations_eval = affiliations_eval['affiliations'].value_counts().reset_index()
affiliations_eval.columns = ['affiliations', 'count']
affiliations_eval.to_csv('affiliations_list_eval.csv', index=False)

contributor_eval = pd.DataFrame(np.array(contributors_all))
contributor_eval.columns = ['contributors']
contributor_eval = contributor_eval['contributors'].value_counts().reset_index()
contributor_eval.columns = ['contributors', 'count']
contributor_eval.to_csv('contributors_list_eval.csv', index=False)
```
